```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


# Ensure the required columns exist
if "Activity" not in mitigation_df.columns or "Description" not in mitigation_df.columns:
    raise ValueError("Excel file must contain 'Activity' and 'Description' columns.")

# Extract activities and descriptions
activities = mitigation_df["Activity"].dropna().tolist()
descriptions = mitigation_df["Description"].dropna().tolist()

# Step 1: Find the best-matching activity using TF-IDF + Cosine Similarity
vectorizer = TfidfVectorizer()
activity_vectors = vectorizer.fit_transform([contract_text] + activities)  # First item is contract, rest are activities
similarities = cosine_similarity(activity_vectors[0:1], activity_vectors[1:])[0]  # Compare contract to all activities

best_activity_idx = similarities.argmax()
best_activity_score = similarities[best_activity_idx] * 100  # Convert to %

best_activity_text = activities[best_activity_idx]

# Step 2: Compare contract with the chosen activity's description
description_vectors = vectorizer.fit_transform([contract_text, descriptions[best_activity_idx]])
description_similarity = cosine_similarity(description_vectors[0:1], description_vectors[1:])[0][0] * 100  # Convert to %

# Print Results
print(f"‚úÖ Activity Match Score: {best_activity_score:.2f}%")
print(f"üîπ Best Matching Activity: {best_activity_text}")
print(f"‚úÖ Description Match Score: {description_similarity:.2f}%")
print(f"üîπ Closest Description: {descriptions[best_activity_idx]}")
```

```python
metric_thresholds = mitigation_df["Metric & Threshold"].dropna().tolist()
if best_activity_score > 50:
    # Extract the corresponding "Metrics & Threshold" for the best-matching activity
    best_metric_threshold = metric_thresholds[best_activity_idx]

    # Step 4: Compare contract with the chosen activity's "Metrics & Threshold"
    metric_threshold_vectors = vectorizer.fit_transform([contract_text, best_metric_threshold])
    metric_threshold_similarity = cosine_similarity(metric_threshold_vectors[0:1], metric_threshold_vectors[1:])[0][0] * 100  # Convert to %

    # Print Results for Metrics & Threshold
    print(f"‚úÖ Metrics & Threshold Match Score: {metric_threshold_similarity:.2f}%")
    print(f"üîπ Closest Metrics & Threshold: {best_metric_threshold}")
else:
    print("‚ùå Activity Match Score is below 50%. Skipping Metrics & Threshold alignment.")

```
